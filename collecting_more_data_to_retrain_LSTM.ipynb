```python
import yt_dlp
import os

output_dir = "/content/drive/MyDrive/shoplifting/youtube_scrape"
os.makedirs(output_dir, exist_ok=True)

queries = [
    "shoplifting caught on camera CCTV",
    "retail theft security camera footage",
    "caught shoplifting store CCTV",
    "shoplifting compilation surveillance",
    "supermarket theft caught camera"
]

ydl_opts = {
    'format': 'mp4/best[height<=480]',
    'outtmpl': f'{output_dir}/%(title)s.%(ext)s',
    'max_downloads': 15,
    'noplaylist': True,
    'quiet': False,
    'ignoreerrors': True,  # skip errors and continue
}

for query in queries:
    print(f"\nSearching: {query}")
    try:
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            ydl.download([f"ytsearch15:{query}"])
    except Exception as e:
        print(f"  Moving to next query...")
        continue

print("\nAll queries done. Check Drive.")
```

```python
import os
import cv2

output_dir = "/content/drive/MyDrive/shoplifting/youtube_scrape"
files = [f for f in os.listdir(output_dir) if f.endswith('.mp4')]

short = 0  # under 2 min
medium = 0  # 2-10 min
long = 0  # over 10 min

for f in files:
    cap = cv2.VideoCapture(os.path.join(output_dir, f))
    fps = cap.get(cv2.CAP_PROP_FPS)
    frames = cap.get(cv2.CAP_PROP_FRAME_COUNT)
    cap.release()
    if fps > 0:
        duration = frames / fps / 60  # in minutes
        if duration < 2:
            short += 1
        elif duration < 10:
            medium += 1
        else:
            long += 1

print(f"Short (<2 min):   {short}")
print(f"Medium (2-10 min): {medium}")
print(f"Long (>10 min):   {long}")
print(f"Total: {short+medium+long}")
```
